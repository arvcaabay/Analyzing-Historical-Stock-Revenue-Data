```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue#:~:text=Tesla%20revenue%20for%20the%20twelve%20months%20ending%20June,2022%20was%20%2481.462B%2C%20a%2051.35%25%20increase%20from%202021.'

# Set the user-agent header
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with the headers
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Locate the table containing revenue data (adjust as per the HTML structure)
    revenue_table = soup.find('table', {'class': 'historical_data_table'})

    # Extract data from the table and create a DataFrame
    tesla_revenue = pd.read_html(str(revenue_table))[0]

    # Display the last five rows of the dataframe
    print(tesla_revenue.tail())
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue'

# Set the user-agent header
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Make the request with the headers
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the element containing the revenue data (adjust as per the HTML structure)
    revenue_data_element = soup.find('div', {'class': 'revenue-data'})

    # Check if the element is found before trying to get text
    if revenue_data_element:
        # Extract the text containing revenue data
        revenue_data_text = revenue_data_element.get_text()
        
        # Print the last five characters of the extracted text (assuming it contains numerical data)
        print(revenue_data_text[-5:])
    else:
        print("No revenue data element found on the page.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Assuming you have a function make_graph that takes a dataframe and a title and plots the graph
def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price ($)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Extract Tesla stock data using yfinance
tesla_data = yf.download('TSLA', start='2022-01-01', end='2023-01-01')

# Reset the index
tesla_data_reset = tesla_data.reset_index()

# Save the dataframe to a CSV file (optional)
tesla_data_reset.to_csv('tesla_stock_data.csv', index=False)

# Display the first five rows of the dataframe (optional)
print(tesla_data_reset.head())

# Plot the Tesla stock graph
make_graph(tesla_data_reset, 'Tesla Stock Price')
```
